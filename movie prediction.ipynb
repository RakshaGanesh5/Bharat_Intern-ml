```python
import gradio as gr
```

```python
def Movie_Names(Rank, Rating, Year, Duration_of_movie, Genere, Description):
    # Create an array with the input features
    x = np.array([Rank, Rating, Year, Duration_of_movie, Genere, Description])

    # Reshape the array to match the input shape expected by the model
    x = x.reshape(1, -1)

    # Make a prediction using the logistic regression model
    prediction = lr.predict(x)

    # Assuming 'lr' is a classifier, you might want to convert the prediction to a label
    label = str(prediction[0])  # Convert the prediction to a string label

    return label
```

```python
app = gr.Interface(
    fn=Movie_Names,
    inputs=[
        gr.inputs.Number(label="Rank"),
        gr.inputs.Number(label="Rating"),
        gr.inputs.Number(label="Year"),
        gr.inputs.Number(label="Duration_of_movie"),
        gr.inputs.Number(label="Genere"),
        gr.inputs.Number(label="Description"),  
    ],
    outputs=gr.outputs.Textbox(),  # Use Textbox for string output
    title="Developing an ML Model for Movie Recommendation",
)
```

```python
app.launch(show_error=True)
```

```python
app.close()
```
